```python
# url addres of amazon page for scraping data

link = 'https://www.amazon.in/O3-Vitamin-Glowing-Complexion-Suitable/product-reviews/B07V2RJ16H/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
# first thing we create a reqst to get the url we want to scrap
# we create reqst here reqst returns the raw content

req = requests.get(link,headers=headers)
```

### To Extract Names of customer

```python
# HTML Tag of Names = <span class="a-profile-name">Customer</span>
```

```python
#SYNTAX:
# df = soup.find_all('parent tag',attrs={'class','extraction part tag'})


#Names = soup.find_all('span',class_="a-profile-name")
Names=soup.findAll('span',{'class','a-profile-name'})
Names
```

```python
# to extract only names without html tags
custnames=[] # creat a list where names will be store

for i in range(0,len(Names)):
    custnames.append(Names[i].get_text())
    
custnames
```

```python
len(custnames) 
# should be 10 coz only 10 reviews are there in our 1st page 
# but we got 12 because of repeation of 1st 2 reviews
```

```python
# now name Abhinav and customer has been occure twice because thes 2 name also present under top positive and critical 
# to remove them
custnames.pop(0)
custnames.pop(1)
custnames
```

```python
len(custnames)
```

### REVIEW TITLE

```python
# HTML Tags for Title ='<span>Too costly</span>'

# '<a data-hook="review-title" class="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold" 
# href="/gp/customer-reviews/R35EBPC8XJGT7A/ref=cm_cr_arp_d_rvw_ttl?ie=UTF8&amp;ASIN=B07V2RJ16H"> <span>Too costly</span>
#</a>'
```

```python
title = soup.findAll('a',class_="review-title")
title
```

```python
# like we did above we will only extract the review title from the whole data we extracted with tags as well

Title=[] # making a list to store the extracted data

for i in range(0,len(title)):
    Title.append(title[i].get_text())
                 
Title
```

```python
# here we extract the content successfully but with '\n' inthe staet and end of every title
# to remove it 

Title[:]= [i.lstrip('\n') for i in Title] # this is to remove from left

Title[:]= [i.rstrip('\n') for i in Title]# this is to remove from right
```

```python
Title
```

```python
len(Title)
```

### RATINGS

```python
# HTML tags for rating = 
# <i data-hook="review-star-rating" class="a-icon a-icon-star a-star-1 review-rating"><span class="a-icon-alt">1.0 out of 5 stars</span></i>


rate = soup.find_all('i', attrs={'class','review-rating'})
rate
```

```python
# again we pass through a loop to remove all extra tags and extract only ratings

Ratings=[]

for i in range(0,len(rate)):
    Ratings.append(rate[i].get_text())
Ratings
```

```python
len(Ratings) # should be 10 coz only 10 reviews are there in our 1st page 
# but we got 12 because of repeation of 1st & 2nd reviews(same case happend in name excration process)
```

```python
# we pop extra or repeated ratings
Ratings.pop(0)
Ratings.pop(1)
Ratings
```

```python
len(Ratings)
```
